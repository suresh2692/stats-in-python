## 2.Pandas

- A DataFrame is a two-dimensional labeled data structure with columns of potentially different types.
- Example: like a spreadsheet or SQL table.
- DataFrames are the most commonly used _pandas objects_.

**numpy**  - handles “rows” ﬁrst. E.g.,data[0] is the ﬁrst row of an array

**pandas** - starts with the columns. E.g.,df['values'][0] is the ﬁrst element of the column 'values'.


```python
import numpy as np
import pandas as pd
t = np.arange(0,10,0.1)
x = np.sin(t)
y = np.cos(t)
df = pd.DataFrame({'Time':t, 'x':x, 'y':y})
df.head()
```

```python
# loc(row) - fetch df by label in index
print(df.loc[:5])

# iloc(row,column) - fetch df by its number 
print(df.iloc[4:10, [0,2]])
```

```python
# Grouping
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame({'Gender': ['f', 'f', 'm', 'f', 'm',
'm', 'f', 'm', 'f', 'm', 'm'],'TV': [3.4, 3.5, 2.6, 4.7, 4.1, 4.1,5.1, 3.9, 3.7, 2.1, 4.3]})

# Group the data
grouped = data.groupby('Gender')
# Do some overview statistics
print(grouped.describe())
# Plot the data:
grouped.boxplot()
plt.show()
#-------------------------------------------#
#Get the groups as DataFrames
df_female = grouped.get_group('f')
# Get the corresponding numpy-array
values_female = grouped.get_group('f').values
```

## 4.1 Univariate plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import seaborn as sns

# Generate the data
x = np.random.randn(500)
```

```python
# Plot-command start --------------------
plt.plot(x,'.')

#or plt.scatter(np.arange(len(x)), x)

# Show plot
plt.show()
```

### Histograms
Histograms to plot the distribution of data. Few **disadvantages** are :-
- They are discontinuous, we cannot come to a conclusion looking at last 100 row's distribution.
- The shape depends on the bin size, If values fall on same bin the bar is stacked up.

    Ex:- x contains [-1,0.15,0.17,0.14,0.2], then values from 2 to 4 will fall in same bin which inturn increases the bar height.

```python
plt.hist(x,bins=25)
plt.show()
```

### Kernel-Density-Estimation (KDE) 
To overcome the problem of histogram, we use **KDE**. This plots a smooth probablity density

```python
#sns.kdeplot(x)
```

### Rug plot
Every data entry is visualized by small vertical ticks

```python
sns.rugplot(x)
```

### cummulative frequencies
A cumulative frequency curve indicates the number (or percent) of data with less than a given value.

- for example when we want to know the data range containing 95 % of all the values.
- for comparing the distribution of values in two or more different groups of individuals

```python
numbins = 25
plt.plot(stats.cumfreq(x,numbins)[0])
```

### Error-Bars
Error-bars are a common way to show mean value and variability when comparing measurement values. 

```python
index = np.arange(5)
y = index**2
errorBar = index/2 # just for demonstration
plt.errorbar(index,y, yerr=errorBar, fmt='o',
capsize=5, capthick=3)
```

### Box Plots
Boxplots are frequently used in scientiﬁc publications to indicate values in two or more groups. 
- The bottom of the box indicate the ﬁrst quartile
- The top of the box indicate the third quartile and
- The line inside the box shows the median

data which lie more than _1.5 * IQR_ (interquartile-range) are **outliers **


```python
plt.boxplot(x, sym='*')
plt.show()
```

### Violin Plots
Box plots combined with the Kde plots form a violin plot, The vertical axis is the same as for the box-plot, but in addition a KDE-plot is shown symmetrically along the horizontal direction 

```python
# Generate the data
nd = stats.norm
data = nd.rvs(size=(100))
nd2 = stats.norm(loc = 3, scale = 1.5)
data2 = nd2.rvs(size=(100))
# Use pandas and the seaborn package
# for the violin plot
df = pd.DataFrame({'Girls':data, 'Boys':data2})
sns.violinplot(df)
```
